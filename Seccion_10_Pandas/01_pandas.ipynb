```python
import pandas as pd
```

Para obtener los datos de un archivos se usa lo siguiente:
Obtener datos de un archivo csv

```python
dFrame = pd.read_csv("supermarkets.csv")
dFrame
```

Obtener datos de un archivo json

```python
dFrame_json = pd.read_json("supermarkets.json")
dFrame_json
```

Obtener datos de un archivo excel

```python
dFrame_excel = pd.read_excel("supermarkets.xlsx")
dFrame_excel
```

Obtener datos de archivo planos

```python
dFrame_txt = pd.read_csv("supermarkets-commas.txt")
dFrame_txt
```

```python
dFrame_txt_semi = pd.read_csv("supermarkets-semi-colons.txt", delimiter=";")
dFrame_txt_semi
```
