```python
import pandas as pd
```

```python
mi_frame = pd.read_csv("supermarkets.csv")
mi_frame
```

Agregando una columna nueva:
Usando la funcion shape()

```python
mi_frame["Continent"] = mi_frame.shape[0]*["North America"]
mi_frame
```

Otra forma de crear un campo nuevo y concatenando 2 columnas existentes en la nueva columna

```python
mi_frame["Country-Continent"] = mi_frame["Country"] + ", " + mi_frame["Continent"]
mi_frame
```

Si queremos transponer el Dataframe, intercambiar filas y columnas que quiere decir que las filas pasan a ser columnas y las columnas pasan a ser filas.
Para ello se haria lo siguiente:

```python
new_dataframe = mi_frame.T
new_dataframe
```

Para agregar una nueva columna a un dataframe transpuesto, se puede usar la misma sintasix que se usa para un Dataframe regular.

```python
new_dataframe[6] = [7, "451 St Bear", "Sin City", "DF 221", "USA", "MaltHan", 12, "Nort America", "USA, North America"]
new_dataframe
```

Y tambien podemos transponer nuevamente ese new_dataframe, usando la misma funcion:

```python
mi_frame = new_dataframe.T
mi_frame
```
